```python
def check_input(element):
    return isinstance(element, str)


def get_substring(s):

    if not s:
        return ""

    char_index = {}
    start = 0
    max_length = 0
    result_start = 0  # Начало самой длинной подстроки

    for i, char in enumerate(s):
        # Если символ уже встречался и его последнее вхождение >= start
        if char in char_index and char_index[char] >= start:
            # Сдвигаем начало подстроки
            start = char_index[char] + 1

        # Обновляем последнее вхождение символа
        char_index[char] = i

        # Проверяем длину текущей подстроки
        current_length = i - start + 1
        if current_length > max_length:
            max_length = current_length
            result_start = start

    # Возвращаем найденную подстроку
    return s[result_start:result_start + max_length]


def main(lst):

    # Ищем самую длинную строку
    longest_str_index = -1
    longest_str = ""

    for i, element in enumerate(lst):
        if check_input(element):
            if len(element) > len(longest_str):
                longest_str = element
                longest_str_index = i

    # Если строк не найдено
    if longest_str_index == -1:
        print("-1, ''")
        return

    # Находим самую длинную подстроку без повторений
    substring = get_substring(longest_str)

    # Выводим результат
    print(f"{longest_str_index}, '{substring}'")
```
